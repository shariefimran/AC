# Pair Plot

```python
import pandas as pd

import matplotlib.pyplot as plt;

import seaborn as sb


iris=pd.read_csv("C:\\Users\\Mekakris\\Documents\\AC\\iris.csv")

sb.set_style('whitegrid')

sb.pairplot(iris,hue="species",size=3)

plt.show()



```

# Univariate analysis using PDF EDA

```python
import pandas as pd

import matplotlib.pyplot as plt;

import seaborn as sns

# lines are called probaility density functions PDF which is smooth form of histogram
#histrogram->x,y plot 
sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"sepal_width")\
    .add_legend()
plt.show()

sns.FacetGrid(iris,hue="species",size=5)\
    .map(plt.scatter,"sepal_width","sepal_length")\
    .add_legend()
plt.show()

```

```python

sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"sepal_length")\
    .add_legend()
    
plt.show()


sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"petal_length")\
    .add_legend()
    
plt.show()


sns.FacetGrid(iris,hue="species",size=5)\
    .map(sns.distplot,"petal_width")\
    .add_legend()
    
plt.show()

# pick one feature which is giving more information
```
