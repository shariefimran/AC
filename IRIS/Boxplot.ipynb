# BoxPlot

```python
import seaborn as sns
import pandas as pd

iris=pd.read_csv("C:\\Users\\Mekakris\\Documents\\AC\\iris.csv")

sns.boxplot(x='species',y='petal_length',data=iris)

# this gives the % of overlapping whic leads to error.
#In box each line gives percentile
# Lines are called whiskers
```

# Violin plot

Combination of both Boxplot and Histograms

```python


sns.violinplot(x='species',y='petal_length',data=iris)

# this gives the % of overlapping whic leads to error.
#In box each line gives percentile
# Lines are called whiskers
```
