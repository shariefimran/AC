```python
import matplotlib.pyplot as plt

plt.plot([20,25,70,91,])  # this is taken as y and as no x is provided index are taken as x axis
plt.xlabel('age')
plt.ylabel('indices')
plt.title('age Index')
plt.show()
```

```python
# another example with
plt.plot(['kris','srava','test'],[25,72,29])
plt.xlabel('names')
plt.ylabel('age')
plt.grid()

plt.show()

```

```python
# another example with
plt.plot(['kris','srava','test'],[25,72,29],'go') #r->color of dots are redn and  o->should be dots not line
plt.xlabel('names')
plt.ylabel('age')
plt.grid()

plt.show()

```

# Advanced Plotting


```python
import numpy as np

a=np.arange(0,1,.1)

print(a)

plt.plot(a,a**5,'b--',label='^5') # b for blue
plt.plot(a,a**6,'rs',label='^6') #red and s for square
plt.plot(a,a**7,'g^',label='^7') # ^ for triangle

plt.legend()
plt.grid()

plt.show()
```

# Plot with params

```python
 a=[1,3,4,4]
b=[2,4,5,6]

plt.plot(a,b,linewidth=2)
plt.show()
```

# Plot with set params

```python
x=list(x**10 for x in  [2,3,4])
a=list(x**2 for x in  [2,3,4])
y=list(x**4 for x in  [1,3,4])
z=list(x**6 for x in  [1,3,4])

line=plt.plot(x,a,y,z) # this will plot 2 lines
plt.setp(line[0],color='r',linewidth=1)
plt.setp(line[1],color='b',linewidth=1)
plt.grid()
plt.show()
```

# Subplot

```python
plt.figure(1)

plt.subplot(211) # 2 rows 1 column and 1st section
plt.plot([1,2,3])

plt.subplot(212)

plt.plot([2,5,6])
plt.show()


plt.figure(2)

plt.subplot(211) # 2 rows 1 column and 1st section
plt.plot([12,2,3])

plt.subplot(212)

plt.plot([2,5,6])
plt.show()
```
