```python
import matplotlib.pyplot as plt

plt.plot([20,25,70,91,])  # this is taken as y and as no x is provided index are taken as x axis
plt.xlabel('age')
plt.ylabel('indices')
plt.title('age Index')
plt.show()
```

```python
# another example with
plt.plot(['kris','srava','test'],[25,72,29])
plt.xlabel('names')
plt.ylabel('age')
plt.grid()

plt.show()

```

```python
# another example with
plt.plot(['kris','srava','test'],[25,72,29],'ro') #r->color of dots are redn and  o->should be dots not line
plt.xlabel('names')
plt.ylabel('age')
plt.grid()

plt.show()

```

# Advanced Plotting


```python
import numpy as np

a=np.arange(0,1,.1)

plt.plot(a,a**5,'b--',label='^5') # b for blue
plt.plot(a,a**6,'rs',label='^6') #red and s for square
plt.plot(a,a**7,'g^',label='^7') # ^ for triangle

plt.legend()
plt.grid()

plt.show()
```

# Plot with params

```python
 a=[1,3,4,4]
b=[2,4,5,6]

plt.plot(a,b,linewidth=2)
plt.show()
```

# Plot with set params

```python
x=[1,3,4]
a=[1,3,4]**2
y=[1,3,4]**4
z=[1,3,4]**5

line=plt.plot(x,a,y,z) # this will plot 2 lines

plt.setp(line[0])
```
