# DB Scan

Density Based Clustering -> DB Scan

How to measure Density? 

Whole idea is divide points  into dense and sparse regions

# Measuring Density

## Min points and EPS

 Density at p -> number of points within a hyper sphere(circle in 2d) of radius EPS around p

 Dense Region-> Circle of radius EPS that contains atleast MIN Points

## Core Point;Border  Point;Noise Point     

    Core Point-> If P has > Min points of EPS then its a core point. This is always part of Dense Region
    
    Border  Point-> P has < Min POints of ESP but lies in range of Core Point
    
    Noise Point -> Neither Core/Border Point
    
# Density Edge and Density Connected  Points

Density Edge (P,Q) -> Both are core points and distance b/w them is <EPS then its Density Edge
    Density Connected  Points-> (p1,p2,p3,p4) where P1->P2,P2-P3,P3->P4 are denstiy edges then P1,P4 then density connected edges
    
# DB Scan Algorithm

    ->  Foreach point Xi label Core/Border/Noise point(done by range queries or use brute force)
    ->  Remove all noise points as they belong to sparse matrix
    ->  For each core point 
           -> Create Cluster
           -> Add all points that are density connected to this point to this cluster
           -> Assign border point to nearst Core point cluster

# Hyper Parameters

## Min Points
    Min Points-> +. d+1 Where d is dimensonality of dataset
    Typially Min Points = 2* dimensions and for a noisy dataset selet big Min Points
    Gnerally domain expert determine the Min POints

## Right EPS
    -> Nth EPS from point (in sorted order)
    -> Elbow Knee method to determine the noisy point

# Advantages/Limitations of DB Scan

## Advantages
        -> Resistant to Noise
        -> No need to specify the clusters
        -> Only 2 parameters MIN points and EPS
## Limitations
      -> Dependant on EPS
      -> Not suitable for high dimension because of the Curse of dimensonality
 
 Time Complexity -> O(n log(n))
 Space Complexity-> O(n)
